```python
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score
from sklearn.metrics import silhouette_score
```

- Загрузите набор данных Iris с помощью функции load_iris из scikit-learn. 
- Примените StandardScaler для масштабирования данных.
-Создайте экземпляр модели KMeans, установив количество кластеров равным 3, и обучите его на отмасштабированных данных. 

```python
# Загрузка датасета
iris = load_iris()


scaler = StandardScaler()
scaler.fit(iris.data)
X_scaled = scaler.transform(iris.data)
km = KMeans(n_clusters=3)
km.fit(X_scaled)
km.predict(X_scaled)
print(adjusted_rand_score(iris.target,km.predict(X_scaled)))
print(silhouette_score(X_scaled,km.predict(X_scaled)))
```

Оцените качество полученной модели, используя метрики ARI и силуэтный коэффициент.

```python
import matplotlib.pyplot as plt
import mglearn
plt.figure(figsize=(8, 8))
mglearn.discrete_scatter(X_scaled[:,0],X_scaled[:,1],km.predict(X_scaled))
plt.legend(iris.target_names,loc = "best")
plt.axes().set_title("KMeans - ARI:{0:.2f}\n Silhouette:{1:.2f}".format(adjusted_rand_score(iris.target,km.predict(X_scaled)),silhouette_score(X_scaled,km.predict(X_scaled))))
```

# Агломеративная кластеризация

```python
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import adjusted_rand_score
from sklearn.metrics import silhouette_score
```

- Загрузите набор данных Iris с помощью функции load_iris из scikit-learn. 
- Примените StandardScaler для масштабирования данных.
- Создайте экземпляр модели AgglomerativeClustering, установив количество кластеров равным 3, и обучите его на отмасштабированных данных.

```python
# Загрузка датасета
iris = load_iris()


scaler = StandardScaler()
scaler.fit(iris.data)
X_scaled = scaler.transform(iris.data)
ag = AgglomerativeClustering(n_clusters=3)
assigment = ag.fit_predict(X_scaled)

print(adjusted_rand_score(iris.target,assigment))
print(silhouette_score(X_scaled,assigment))
```

Оцените качество полученной модели, используя метрики ARI и силуэтный коэффициент.

```python
import matplotlib.pyplot as plt
import mglearn
plt.figure(figsize=(8, 8))
mglearn.discrete_scatter(X_scaled[:,0],X_scaled[:,1],assigment)
plt.legend(iris.target_names,loc = "best")
plt.axes().set_title("KMeans - ARI:{0:.2f}\n Silhouette:{1:.2f}".format(adjusted_rand_score(iris.target,assigment),silhouette_score(X_scaled,assigment)))
```

# DBSCAN

```python
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
from sklearn.metrics import adjusted_rand_score
from sklearn.metrics import silhouette_score
```

- Загрузите набор данных Iris с помощью функции load_iris из scikit-learn. 
- Примените StandardScaler для масштабирования данных.
- Обучите модель DBSCAN, используя значения параметра min_samples равные 1, 2, … 5  и значения параметра eps равные 0,5, 1, … 2,5. 

```python
import matplotlib.pyplot as plt
import mglearn

# Загрузка датасета
iris = load_iris()

samples=[]
epsilons = []
count_cl=[]
ari=[]

ans = []
min_samp = [1,2,3,4,5]
epsi = [0.5, 1, 1.5, 2, 2.5]

scaler = StandardScaler()
scaler.fit(iris.data)
X_scaled = scaler.transform(iris.data)
i = 1
for s in min_samp:
  for e in epsi:
    samples.append(s)
    epsilons.append(e)
    db = DBSCAN(min_samples = s,eps = e)
    assigment = db.fit_predict(X_scaled)
    ans.append(assigment)
    plt.figure(figsize=(8, 8))
    mglearn.discrete_scatter(X_scaled[:,0],X_scaled[:,1],assigment)
    plt.legend(iris.target_names,loc = "best")
    #plt.axes().set_title(f"{}")
    count_cl.append(len(set(assigment)))
    ari.append(adjusted_rand_score(iris.target,assigment))
    try:
      plt.axes().set_title("KMeans - ARI:{0:.2f}\n Silhouette:{1:.2f}\n Num:{2:.0f}".format(adjusted_rand_score(iris.target,assigment),silhouette_score(X_scaled,assigment),len(set(assigment))))

      #plt.axes().set_title("KMeans - ARI:{0:.2f}\n Silhouette:{1:.2f}\n Num:{}".format(adjusted_rand_score(iris.target,assigment),silhouette_score(X_scaled,assigment)))
    except ValueError:
      plt.axes().set_title("KMeans - ARI:{0:.2f}\n Silhouette:{1:.2f}\n Num:{2:.0f}".format(adjusted_rand_score(iris.target,assigment),0.00,len(set(assigment))))
    i+=1
```

```python
import pandas as pd

df = pd.DataFrame()

df['min_sample']=samples
df['eps']=epsilons
df['num_cl']=count_cl
df['ARI']=ari
df
```
